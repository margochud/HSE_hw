Пусть $\xi_i \in \mathcal{NegBin}(r, p)$. Покажите, что при $r \to \inf$ негативно биномиальное распределение $NegBin(r, 1 - \frac{\lambda}{r+\lambda})$ стремится к распределению Пуассона $Pois(\lambda)$.

Негативно биномиальное распределени --- количество неудач в серии ИБ с вероятностью $p$, произошедших до $r$-ого успеха.

$$f(k) = \binom{k+r-1}{r-1}p^r(1-p)^k$$


```python
cmap = plt.get_cmap("tab10").colors

plt.figure(figsize=(15, 15))

lam = 21

for k, r in enumerate([10, 50, 100, 500, 1000, 5000, 10000, 50000, 100000]):
    p = 1 - lam / (lam + r)
    c = [st.nbinom.pmf(i, r, p) for i in range(30)]
    plt.subplot(3, 3, k+1)
    plt.tight_layout()
    y = [st.poisson.pmf(i, lam) for i in range(30)]
    plt.plot(list(range(30)), y, '.-', c='k', label="Poiss")
    plt.scatter(list(range(30)), c, marker='s', color=cmap[k], label="r = " + str(r))
    #plt.xticks(x, x)
    plt.xlabel("k")
    plt.ylabel(r"$P(\xi _n=k)$")
    plt.legend()
plt.show()
```
